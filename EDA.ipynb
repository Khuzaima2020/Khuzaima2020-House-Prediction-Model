```python
#Import libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_main = pd.read_csv('For EDA.csv')
```

```python
df_main.head()
```

```python
df_main.info()
```

```python
#check of outliers
df_main.describe()
```

```python
#visulaize the property type
sns.countplot(df_main['property_type'])
plt.show()
```

```python
#plot bar graph for Province name and price
df_main.groupby('province_name')['price'].mean().plot.bar()
plt.title('Province name vs Price')
plt.xlabel('Province name')
plt.ylabel('Price')
plt.show()
```

```python
#bar plot for city and prices
df_main.groupby('city')['price'].mean().plot.bar()
plt.title('City vs Price')
plt.xlabel('City')
plt.ylabel('Price')
plt.show()
```

```python
#bar plot for price and Location
df_main.groupby('location')['price'].mean().plot.bar()
plt.title('Location vs Price')
plt.xlabel('Location')
plt.ylabel('Price')
plt.show()
```

```python
df_main.head()
```

```python
#price distribution according to number of bath and bedrooms
sns.scatterplot(x='baths',y='bedrooms',hue='price',data=df_main)
plt.title('Price vs Number of Bathrooms and Bedrooms')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Number of Bedrooms')
plt.show()
```

```python
#remove values with bedrooms greater than 30 and baths greater than 50
df_main = df_main[df_main['bedrooms']<30]
df_main = df_main[df_main['baths']<50]

#price distribution according to number of bath and bedrooms
sns.scatterplot(x='baths',y='bedrooms',hue='price',data=df_main)
plt.title('Price vs Number of Bathrooms and Bedrooms')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Number of Bedrooms')
plt.show()
```

```python
df_main.head(2)
```

```python
#sales according to year
df_main['year'].value_counts().plot.bar()
plt.title('Sales according to year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
#sales according to month
df_main['month'].value_counts().plot.bar()
plt.title('Sales according to month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
